## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Car_Purchasing_Data.csv')
df
```

## Dataset Visualization

```python
sns.pairplot(df)
```

## Create Testing and Training Dataset

```python
X = df.drop(['Customer Name', 'Customer e-mail', 'Country', 'Car Purchase Amount'], axis=1)
X
```

```python
y = df['Car Purchase Amount']
y
```

```python
y.shape
```

```python
X.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
scaler.data_max_
```

```python
scaler.data_min_
```

```python
y = y.values.reshape(-1, 1)
```

```python
y_scaled = scaler.fit_transform(y)
y_scaled
```

## Training the Model

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.25)
```

```python
from tensorflow.python import keras
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
#model.add(Dense(25, input_dim=5, activation='relu')) # how many inputs
#model.add(Dense(25, activation='relu'))
#model.add(Dense(1, activation='linear')) # one output
```

```python
model = Sequential()

model.add(Dense(40, input_dim=5, activation='relu'))
model.add(Dense(40, activation='relu'))
model.add(Dense(1, activation='linear'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
epochs_hist = model.fit(X_train, y_train, epochs=100, batch_size=50, verbose=1, validation_split=0.2)
```

## Model Evaluation

```python
epochs_hist.history.keys()
```

```python
plt.plot(epochs_hist.history['loss'])
plt.plot(epochs_hist.history['val_loss'])
plt.title('Model Loss Progress During Training')
plt.ylabel('Training and Validation Loss')
plt.xlabel('Epoch Number')
plt.legend(['Training Loss', 'Validation Loss'])
```

```python
# Picking  out a use case to test Gender, Age, Salary, CC Debt, Net Worth
X_test = np.array([[1, 50, 50000, 10000, 60000]])
y_predict = model.predict(X_test)
```

```python
print(f'Predicted Purchase Amount: {y_predict}')
```
